## Chessboard

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pix = 256
```

```python
plt.figure(figsize=(10,10))
m = np.zeros((int(pix*1.2), int(pix*1.2)))

for i in range(pix):
    
    i_odd = False
    if i//(pix/4) % 2 == 1:
        i_odd = True
        
    i += int(pix*0.1)
    
    for j in range(pix):
        
        j_odd = False
        if j//(pix/4) % 2 == 1:
            j_odd = True
            
        j += int(pix*0.1)
        
        if i_odd and not j_odd or not i_odd and j_odd:
            m[i][j] = 255
        else:
            m[i][j] = 127
            
plt.imshow(m, cmap = "gray")
plt.grid()
plt.show()
```
